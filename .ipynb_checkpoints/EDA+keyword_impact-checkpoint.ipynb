### Users who have already done Sessions vs Users who have not done Sessions 

```python
sessions_count = df.groupby("pais")["sesion_completada"].value_counts().unstack(fill_value=0)
```

```python
sessions_count = sessions_count.sort_values(by="Yes", ascending=False).head(10)
```

```python
plt.figure(figsize=(28, 12))
ax = sessions_count.plot(kind="bar", stacked=True, width=0.9, color=["#FF00FF","#0000FF"])

plt.title("Usuarias que Hicieron Sesión vs No Hicieron Sesión", fontsize=18, pad=20)
plt.ylabel("Número de Usuarias", fontsize=14)
plt.xlabel("País", fontsize=14)

plt.legend(title="Sesión Completada", labels=["No", "Sí"], bbox_to_anchor=(1, 1), loc='best')

for i, (index, row) in enumerate(sessions_count.iterrows()):
    no_session = row["No"]
    yes_session = row["Yes"]
    total = no_session + yes_session
  
    if no_session > 200:
        plt.text(i, no_session/2, f'{int(no_session)}', ha='center', va='center', color='white', fontweight='bold')
  
    if yes_session > 100:
        plt.text(i, no_session + yes_session/2, f'{int(yes_session)}', ha='center', va='center', color='white', fontweight='bold')
    
    plt.text(i, total, f' {int(total)}', ha='center', va='bottom', fontweight='bold', color='black')


plt.ylim(0, sessions_count.sum(axis=1).max() * 1.2)


plt.xticks(rotation=45, ha="right")

plt.show()
```

- We observe that Spain is the main country that did the first session, followed up by Unknown. The rest of sessions are distributed by the other countries (mostly from Latam). Let's group up thess countries for a further analysis. 

```python
def zone_category(country):
    if country == "España":
        return "España"
    elif country == "Desconocido":
        return "Desconodido"
    else:
        return "Otras Zonas"
```

```python
df["categoria_zona"] = df["pais"].apply(zone_category)
```

```python
sessions_count_zone = df.groupby("categoria_zona")["sesion_completada"].value_counts().unstack(fill_value=0)
```

```python
sessions_count_zone = sessions_count_zone.sort_values(by="Yes", ascending=False)
sessions_count_zone
```

- We can try to clear up "Desconocido" group in "categoria_zona" by "zona_horaria"

```python
df.groupby("categoria_zona")["zona_horaria"].value_counts()
```

- There's no way to clear "Desconocido". All fields in that group are incompleted.
- Let's plot the grouped distribution.

```python
plt.figure(figsize=(28, 12))
ax = sessions_count_zone.plot(kind="bar", stacked=True, width=0.9, color=["#FF00FF","#0000FF"])

plt.title("Usuarias que Hicieron Sesión vs No Hicieron Sesión", fontsize=18, pad=20)
plt.ylabel("Número de Usuarias", fontsize=14)
plt.xlabel("Zona", fontsize=14)

plt.legend(title="Sesión Completada", labels=["No", "Sí"], bbox_to_anchor=(1, 1), loc='upper left')

for i, (index, row) in enumerate(sessions_count_zone.iterrows()):
    no_session = row["No"]
    yes_session = row["Yes"]
    total = no_session + yes_session

    if no_session > 0:
        plt.text(i, no_session/2, f'{int(no_session)}', ha='center', va='center', color='white', fontweight='bold')

    if yes_session > 0:
        plt.text(i, no_session + yes_session/2, f'{int(yes_session)}', ha='center', va='center', color='white', fontweight='bold')

plt.ylim(0, sessions_count_zone.sum(axis=1).max() * 1.1)

plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
def zone_sessions_percentage(df):
    grouped = df.groupby("categoria_zona")["sesion_completada"].value_counts(normalize=True).unstack()
    
    percentages = (grouped['Yes'] * 100).round(2)
    
    return percentages.to_dict()
```

```python
sessions_zone_percentage = zone_sessions_percentage(df)
```

```python
sessions_zone_percentage
```

```python
sorted_data = sorted(sessions_zone_percentage.items(), key=lambda x: x[1], reverse=True)
labels, sizes = zip(*sorted_data)

colors = ['#FF00FF', '#0000FF', '#8000FF']

fig, ax = plt.subplots(figsize=(8, 4))
wedges, texts, autotexts = ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', 
                                startangle=90, pctdistance=0.85, textprops={'color': 'white'})

centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

plt.setp(autotexts, size=8, weight="bold")
plt.setp(texts, size=12)

ax.axis('equal')

plt.title("Distribución de Sesiones Completadas por Zona", fontsize=16, pad=20)

plt.legend(wedges, labels, title="Zonas", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.tight_layout()
plt.show()
```

- After plotting the grouped by zones (Spain, Unknown, Other Zones) we could see the distribution and percentage of completed sessions. It's clear that Spain is most effective target with positive response for the Marketing actions.
- In the Yes/No distribution, Spain is the only zone with Yes overcoming No

### Inputs cleaning to tags

```python
# NLTK will help us to earse unnecesary prepositions and others

stop_words = list(stopwords.words("spanish"))
```

```python
stop_words.extend(["según", "ahora", "después", "todas", "toda", "todo", "todos", "sólo", "solo", "sido", "están",
                   "estan", "hacer", "hecho", "puede", "tras", "cabe", "bajo", "durante", "mediante", "cada", "me",
                  "lunes", "martes", "miércoles", "jueves", "viernes", "sabado", "domingo"])
```

```python
def text_cleaner(text):
    words = []
    for word in gensim.utils.simple_preprocess(text):
        if word not in gensim.parsing.preprocessing.STOPWORDS and word not in stop_words and len(word) > 4:
            words.append(word)
    return words
```

```python
problema_principal = df["problema_principal"].apply(text_cleaner)
```

```python
cambios_deseados = df["cambios_deseados"].apply(text_cleaner)
```

```python
busqueda_en_terapia = df["busqueda_en_terapia"].apply(text_cleaner)
```
